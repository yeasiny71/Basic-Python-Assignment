# Label Encoder using manual approach

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['sex'] = le.fit_transform(df['sex'])
```

```python
df['smoker'] = le.fit_transform(df['smoker'])
```

```python
df['region'] = le.fit_transform(df['region'])
```

# Label encoder using for loop

```python
df1.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

# One hot Encoder using manual approach

```python
df2.head()
```

```python
dummy = pd.get_dummies(df2['sex'],drop_first=True,prefix='gender')   #df2['sex'] = df2.sex
```

```python
dummy.head()
```

```python
dummy1 = pd.get_dummies(df2.smoker,drop_first=True,prefix='smoker')
dummy1.head()
```

```python
dummy2 = pd.get_dummies(df2['region'],drop_first=True,prefix='region')
dummy2.head()
```

```python
df2.drop(columns= ['sex','smoker','region'],inplace=True)
df2.head()
```

```python
df2 = pd.concat([df2,dummy,dummy1,dummy2],axis = 1)
```

```python
df2.head()
```

# One hot encoding using for loop

```python
df3.head()
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        dummy3 = pd.get_dummies(df3[col],drop_first=True,prefix = col)
        df3.drop(col,axis=1,inplace = True)
        df3 = pd.concat([df3,dummy3],axis = 1)
df3.head()
        
```

# Ordinal Encoder using manual approach

```python
df4.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sex_order = df4.sex.unique()
sex_order
```

```python
ordinal = OrdinalEncoder(categories=[sex_order])
ordinal
```

```python
encoded = ordinal.fit_transform(df4[['sex']])
encoded
```

```python
df4['sex'] = pd.DataFrame(encoded,columns = ['SEX'])
df4.head()
```

```python
smoker_order = df4.smoker.unique()
smoker_order
```

```python
ordinal = OrdinalEncoder(categories=[smoker_order])
ordinal
```

```python
encoded = ordinal.fit_transform(df4[['smoker']])
encoded
```

```python
df4['smoker'] = pd.DataFrame(encoded,columns=['SMOKER'])
df4.head()
```

```python
region_order = df4.region.unique()
region_order
```

```python
ordinal = OrdinalEncoder(categories=[region_order])
ordinal
```

```python
encoded = ordinal.fit_transform(df4[['region']])
encoded
```

```python
df4['region'] = pd.DataFrame(encoded,columns = ['REGION'])
df4.head()
```

# Ordinal encoder using for loop

```python
df5.head()
```

```python
for col in df5.columns:
    if is_numeric_dtype(df5[col]):
        continue
    else:
        order = df5[col].unique()
        ordinal = OrdinalEncoder(categories=[order])
        encoded = ordinal.fit_transform(df5[[col]])
        df5[col] = pd.DataFrame(encoded,columns=[col])
df5.head()
```
